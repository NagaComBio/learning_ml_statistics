### Importing basic packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Data

```python
X = np.linspace(-np.pi, np.pi, 255, endpoint=True)
S, C = np.sin(X), np.cos(X)
```

### Basic plot

```python
plt.plot(X, S)
plt.plot(X, C)
plt.show()
```

### Changing color and linewidth

```python
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.show()
```

### Setting ticks and labels

```python
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r"$-\pi$", r"$-\pi/2$", 0, r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, 0, 1], 
          [r"$-1$", 0, r"$+1$"])

plt.show()
```

### Moving the spines

I have modifed the spines, by placing the `ax.spines[..]set_position('center')`, I feel this is much simplier.

```python
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_position('center')
ax.yaxis.set_ticks_position('right')

plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r"$-\pi$", r"$-\pi/2$", 0, r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, 0, 1], 
          [r"$-1$", 0, r"$+1$"])

plt.show()
```

### Adding a legend
with latex formatting

```python
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_position('center')
ax.yaxis.set_ticks_position('right')

plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label=r"$\sin(X)$")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label=r"$\cos(X)$")

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r"$-\pi$", r"$-\pi/2$", 0, r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, 0, 1], 
          [r"$-1$", 0, r"$+1$"])

plt.legend(loc='upper left', frameon=False)
plt.show()
```

### Adding an annotation

```python
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_position('center')
ax.yaxis.set_ticks_position('right')

t=(2*np.pi)/3

plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label=r"$\sin(X)$")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label=r"$\cos(X)$")
plt.plot([t, t], [0, np.sin(t)], color="red", linewidth=1.5, linestyle="--")
plt.plot([t, t], [0, np.cos(t)], color="blue", linewidth=1.5, linestyle="--")
plt.scatter(t, np.sin(t), color="red", s=50)
plt.scatter(t, np.cos(t), color="blue", s=50)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r"$-\pi$", r"$-\pi/2$", 0, r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, 0, 1], 
          [r"$-1$", 0, r"$+1$"])

plt.annotate(xy=[t, np.sin(t)], 
             s=r"$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$",
             fontsize=12,
             xytext=[+10,+30],             
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3,rad=.2")
            )

plt.annotate(xy=[t, np.cos(t)], 
             s=r"$\cos(\frac{2\pi}{3})=-\frac{1}{2}$",
             fontsize=12,
             xytext=[-90,-40],             
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3,rad=.2")
            )

plt.legend(loc='upper left', frameon=False)
plt.show()
```

### Adding artistic visuals

To bring the ticks boxes on top of the plot, the plots were moved to -1, -2 positions using `zorder`

```python
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_position('center')
ax.yaxis.set_ticks_position('right')

t=(2*np.pi)/3

plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label=r"$\sin(X)$", zorder=-1) #zorder here
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label=r"$\cos(X)$", zorder=-2)
plt.plot([t, t], [0, np.sin(t)], color="red", linewidth=1.5, linestyle="--")
plt.plot([t, t], [0, np.cos(t)], color="blue", linewidth=1.5, linestyle="--")
plt.scatter(t, np.sin(t), color="red", s=50)
plt.scatter(t, np.cos(t), color="blue", s=50)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
           [r"$-\pi$", r"$-\pi/2$", 0, r"$\pi/2$", r"$\pi$"])
plt.yticks([-1, 0, 1], 
          [r"$-1$", 0, r"$+1$"])

plt.annotate(xy=[t, np.sin(t)], 
             s=r"$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$",
             fontsize=12,
             xytext=[+10,+30],             
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3,rad=.2")
            )

plt.annotate(xy=[t, np.cos(t)], 
             s=r"$\cos(\frac{2\pi}{3})=-\frac{1}{2}$",
             fontsize=12,
             xytext=[-90,-40],             
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3,rad=.2")
            )

## Updated part
for labels in ax.get_xticklabels() + ax.get_yticklabels():
    labels.set_fontsize(16)
    labels.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
    
###
ax.locator_params
plt.legend(loc='upper left', frameon=False)
plt.show()
```
